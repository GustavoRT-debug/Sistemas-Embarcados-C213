```python
#8
import tkinter as tk
from tkinter import Label, Entry, Button
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
import control as cnt

# Variáveis globais
kp = 0
ti = 0
td = 0
setpoint = 0

# Função para atualizar as variáveis globais a partir dos campos de entrada
def atualizar_variaveis_globais():
    global kp, ti, td, setpoint
    kp = float(entrada_kp.get())
    ti = float(entrada_ti.get())
    td = float(entrada_td.get())
    setpoint = float(entrada_setpoint.get())

    # Atualizar os rótulos com os valores calculados
    label_kp_valor.config(text=f'KP: {kp:.2f}')
    label_ti_valor.config(text=f'TI: {ti:.2f}')
    label_td_valor.config(text=f'TD: {td:.2f}')

# Função para calcular a resposta do sistema com base nos parâmetros do PID e Setpoint
def calcular_resposta_pid():
    atualizar_variaveis_globais()

    # Cálculo da resposta do sistema de controle
    t = np.linspace(0, 30, 1000)
    num = np.array([kp])
    den = np.array([ti * td, ti, 1, 0])
    H = cnt.tf(num, den)

    (t, y) = cnt.step_response(H, t)

    # Plot da curva da resposta do sistema
    ax.clear()
    ax.plot(t, y)
    ax.set_xlabel('Tempo (s)')
    ax.set_ylabel('Amplitude')
    ax.set_title('Resposta do Sistema')
    canvas.draw()

# Função para calcular os parâmetros do PID usando o método CHR
def calcular_metodo_chr():
    global kp, ti, td
    tau = 16.6 / 4
    theta = 1.9

    kp = ((0.95 * tau) / (setpoint * theta))
    ti = (1.357 * tau)
    td = (0.473 * theta)

    calcular_resposta_pid()

    # Atualiza os rótulos com os valores calculados
    label_kp_valor.config(text=f'KP: {kp:.2f}')
    label_ti_valor.config(text=f'TI: {ti:.2f}')
    label_td_valor.config(text=f'TD: {td:.2f}')

# Função para calcular os parâmetros do PID usando o método da Integral
def calcular_metodo_integral():
    global kp, ti, td
    tau = 16.6 / 4
    theta = 1.9

    kp = 1 / ((theta / tau + 0.2) / setpoint)
    ti = ((0.3 * (theta / tau) + 1.2) / (theta / tau + 0.08)) * theta
    td = (1 / (90 * (theta / tau))) * theta

    calcular_resposta_pid()

    # Atualizar os rótulos com os valores calculados
    label_kp_valor.config(text=f'KP: {kp:.2f}')
    label_ti_valor.config(text=f'TI: {ti:.2f}')
    label_td_valor.config(text=f'TD: {td:.2f}')

# Crie a janela principal do aplicativo
root = tk.Tk()
root.title("Interface PID")

# Rótulos e campos de entrada para os parâmetros do PID e Setpoint
Label(root, text="KP").grid(row=0, column=0)
entrada_kp = Entry(root)
entrada_kp.grid(row=0, column=1)

Label(root, text="TI").grid(row=1, column=0)
entrada_ti = Entry(root)
entrada_ti.grid(row=1, column=1)

Label(root, text="TD").grid(row=2, column=0)
entrada_td = Entry(root)
entrada_td.grid(row=2, column=1)

Label(root, text="Setpoint").grid(row=3, column=0)
entrada_setpoint = Entry(root)
entrada_setpoint.grid(row=3, column=1)

# Rótulos para exibir os valores de KP, TI e TD
label_kp_valor = Label(root, text=f'KP: {kp:.2f}')
label_kp_valor.grid(row=0, column=2)

label_ti_valor = Label(root, text=f'TI: {ti:.2f}')
label_ti_valor.grid(row=1, column=2)

label_td_valor = Label(root, text=f'TD: {td:.2f}')
label_td_valor.grid(row=2, column=2)

# Botões para calcular a resposta com diferentes métodos
botao_metodo_chr = Button(root, text="Calcular (Método CHR)", command=calcular_metodo_chr)
botao_metodo_chr.grid(row=4, column=0)

botao_metodo_integral = Button(root, text="Calcular (Método da Integral)", command=calcular_metodo_integral)
botao_metodo_integral.grid(row=4, column=1)

# Botão para calcular a resposta do sistema
botao_calcular = Button(root, text="Calcular Resposta", command=calcular_resposta_pid)
botao_calcular.grid(row=5, column=0, columnspan=3)

# Configuração do gráfico Matplotlib
fig, ax = plt.subplots(figsize=(6, 4))
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().grid(row=6, column=0, columnspan=3)
canvas.draw()

root.mainloop()



```
