```python
#4
num = np.array([k])  # Numerador (coeficiente associado ao zero)
den = np.array([tau, 1])  # Denominador (coeficientes associados aos polos)
H = cnt.tf(num, den)  # Cria a função de transferência H
(num_pade, den_pade) = cnt.pade(theta, n_pade)  # Cálculo de Pade
H_pade = cnt.tf(num_pade, den_pade)  # Criação da função de transferência com Pade
Hs = cnt.series(H, H_pade)  # Combinação de H e H_pade
Hmf = cnt.feedback(Hs, 1)  # Criação da malha fechada
(t, y) = cnt.step_response(7 * Hmf, t)  # Simulação da resposta
plt.plot(t, y)
plt.xlabel('t [s]')
plt.ylabel('Amplitude')
plt.title('Malha Fechada')

plt.grid(True)
plt.show()
```

```python
#5
# CHR com 20% de Sobrevalor (Problema Servo -PID)
k = 0.5 #muda
tau = 16.6 /4
theta = 1.9

kp = ((0.95*tau)/(k*theta))
Ti = (1.357*tau)
Td = (0.473*theta)

# FUNC TRANF PLANTA
num = np. array([k])
den = np. array([tau, 1])
H = cnt.tf(num, den)
n_pade = 20
(num_pade, den_pade) = cnt.pade(theta, n_pade)
H_pade = cnt.tf(num_pade, den_pade)
Hs = cnt.series(H, H_pade)

# Controlador proporcional
numkp = np. array ([kp])
denkp = np. array ([1])

#integral
numki = np. array ([kp])
denki = np. array ([Ti,0])

#derivativo
numkd = np. array ([kp*Td,0])
denkd = np. array ([1])

#Construindo o controlador PID
Hkp = cnt.tf(numkp , denkp)
Hki=cnt.tf(numki , denki)
Hkd=cnt.tf(numkd , denkd)
Hctrl1 = cnt.parallel (Hkp , Hki)
Hctrl = cnt.parallel (Hctrl1 , Hkd)
Hdel = cnt.series (Hs , Hctrl)

#Fazendo a realimentação
Hcl = cnt.feedback(Hdel, 1)
t = np . linspace(0, 30, 100)
(t, y) = cnt.step_response(7 * Hcl, t)
plt.plot(t, y)
plt.xlabel('t[s]')
plt.ylabel('Amplitude')
plt.title('Controle PID')


plt.grid()
plt.show()
```
